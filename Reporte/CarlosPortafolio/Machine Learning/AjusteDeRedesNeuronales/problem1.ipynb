```python
import numpy as np
import pandas as pd
from keras import Sequential
from sklearn.model_selection import KFold
from keras.layers import Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
dataframe=pd.read_csv("data/problem1/crime_data.csv")
print(dataframe)
```

```python
x=dataframe[['M','W','H','S']]
x
```

```python
y=dataframe['MR']
y
```

<h3>1. Evalúa con validación cruzada un modelo perceptrón multicapa para las variables que se te asignaron para este ejercicio.</h3>

```python
#Number of columns of X
features = x.shape[1]
features
```

```python
n_folds=5
kfold = KFold(n_splits=n_folds,shuffle=True)
cv_y_test = []
cv_y_pred = []
it=0
for train_index, test_index in kfold.split(x,y):
    it=it+1
    print("it: ",it)
    
    x_train=x[train_index,:]
    y_train=y[train_index]
    
    x_test=x[test_index,:]
    y_test=y[test_index]
    
    #Training Phase
    #Define the MLP Model
    mlp_model = Sequential()
    mlp_model.add(Dense(64, activation='relu', input_dim=features))
    mlp_model.add(Dense(32, activation='relu'))
    mlp_model.add(Dense(16, activation='relu'))
    mlp_model.add(Dense(1))  # Output layer
    
    # Compile the model
    mlp_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
    
    # Train the model
    mlp_model.fit(x_train, y_train, epochs=500, batch_size=32, verbose=0)
    
    # Predict on the test data
    y_pred = mlp_model.predict(x_test)  # No need for np.argmax in regression
    
    # Append the true and predicted values
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)
    
# Convert list of arrays to one array for final evaluation
cv_y_test = np.concatenate(cv_y_test)
cv_y_pred = np.concatenate(cv_y_pred)

mse = mean_squared_error(cv_y_test, cv_y_pred)
mae = mean_absolute_error(cv_y_test, cv_y_pred)
r2 = r2_score(cv_y_test, cv_y_pred)

print(f'MSE: {mse}, MAE: {mae}, R²: {r2}')
```

```python
n_folds=5
kf = KFold(n_splits=n_folds, shuffle = True)
cv_y_test = []
cv_y_pred = []
it=0
for train_index, test_index in kf.split(x, y):
    it += 1
    print("it:", it)
    
    x_train = x[train_index, :]
    y_train = y[train_index]
    
    x_test = x[test_index, :]
    y_test = y[test_index]
    
    # Create and fit the Linear Regression model
    regressor = LinearRegression()
    regressor.fit(x_train, y_train)
    
    # Predict on the test data
    y_pred = regressor.predict(x_test)  # No need for np.argmax in regression
    
    # Append the true and predicted values
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)
    
# Convert list of arrays to one array for final evaluation
cv_y_test = np.concatenate(cv_y_test)
cv_y_pred = np.concatenate(cv_y_pred)

# Calculate performance metrics
mse = mean_squared_error(cv_y_test, cv_y_pred)
mae = mean_absolute_error(cv_y_test, cv_y_pred)
r2 = r2_score(cv_y_test, cv_y_pred)

print(f'MSE: {mse}, MAE: {mae}, R²: {r2}')
```
